````python
import requests
import pandas as pd
import io

# Download abstracts and dates separately then merge
print("Downloading abstracts...")
abs_url = "http://data.nber.org/nber_paper_chapter_metadata/tsv/abs.tsv"
date_url = "http://data.nber.org/nber_paper_chapter_metadata/tsv/date.tsv"
title_url = "http://data.nber.org/nber_paper_chapter_metadata/tsv/title.tsv"

df_abs = pd.read_csv(abs_url, sep='\t', on_bad_lines='skip')
print(f"Abstracts loaded: {len(df_abs)} rows")

print("Downloading dates...")
df_date = pd.read_csv(date_url, sep='\t', on_bad_lines='skip')
print(f"Dates loaded: {len(df_date)} rows")

print("Downloading titles...")
df_title = pd.read_csv(title_url, sep='\t', on_bad_lines='skip')
print(f"Titles loaded: {len(df_title)} rows")

# Merge them
df = df_abs.merge(df_date, on='paper').merge(df_title, on='paper')

# Filter to last 15 years
df['issue_date'] = pd.to_datetime(df['issue_date'], errors='coerce')
df = df[df['issue_date'].dt.year >= 2010]

print(f"\nPapers from 2010 onwards: {len(df)}")
print(df.head(3))
````

````python
import requests
import pandas as pd

print("Downloading dates...")
df_date = pd.read_csv("http://data.nber.org/nber_paper_chapter_metadata/tsv/date.tsv", 
                      sep='\t', on_bad_lines='skip', engine='python')
df_date['issue_date'] = pd.to_datetime(df_date['issue_date'], errors='coerce')
df_date = df_date[df_date['issue_date'].dt.year >= 2010]
print(f"Papers from 2010+: {len(df_date)}")

print("Downloading abstracts...")
df_abs = pd.read_csv("http://data.nber.org/nber_paper_chapter_metadata/tsv/abs.tsv", 
                     sep='\t', on_bad_lines='skip', engine='python')
print(f"Abstracts loaded: {len(df_abs)}")

print("Downloading titles...")
df_title = pd.read_csv("http://data.nber.org/nber_paper_chapter_metadata/tsv/title.tsv", 
                       sep='\t', on_bad_lines='skip', engine='python')
print(f"Titles loaded: {len(df_title)}")

# Merge and filter to 2010+
df = df_date.merge(df_abs, on='paper').merge(df_title, on='paper')
df = df.dropna(subset=['abstract'])

print(f"\nFinal dataset: {len(df)} papers with abstracts")
print(df.head(3))
````

````python
import requests
import json
import time
from collections import defaultdict

all_models = defaultdict(int)
batch_size = 10
papers_list = df[['title', 'abstract']].to_dict('records')
total_batches = len(papers_list) // batch_size

print(f"Total papers: {len(papers_list)}")
print(f"Total batches: {total_batches}")
print("Starting extraction...\n")

for i in range(0, len(papers_list), batch_size):
    batch = papers_list[i:i+batch_size]
    batch_num = i // batch_size + 1
    
    abstracts_text = ""
    for j, p in enumerate(batch):
        abstracts_text += f"\nPaper {j+1}: {p['title']}\n{p['abstract']}\n---"
    
    prompt = f"""Read these economics paper abstracts carefully.
Extract ONLY the specific econometric models, statistical methods, and economic frameworks that are explicitly mentioned or used in the research.

Return ONLY a JSON array of strings. No explanation, no markdown, just the array.
Example: ["Difference-in-Differences", "Instrumental Variables", "DSGE Model"]

If no specific models are mentioned, return an empty array: []

Abstracts:
{abstracts_text}

JSON array:"""

    try:
        response = requests.post("", json={
            "model": "gemma3:12b",
            "prompt": prompt,
            "stream": False
        }, timeout=120)
        
        raw = response.json()["response"].strip()
        raw = raw.replace("```json", "").replace("```", "").strip()
        
        # find the array in the response
        start = raw.find('[')
        end = raw.rfind(']') + 1
        if start != -1 and end != 0:
            models = json.loads(raw[start:end])
            for m in models:
                if isinstance(m, str) and len(m) > 2:
                    all_models[m.strip()] += 1
        
        if batch_num % 10 == 0:
            print(f"Batch {batch_num}/{total_batches} — {len(all_models)} unique models found so far")
            
    except Exception as e:
        print(f"Batch {batch_num} failed: {e}")
        continue

print(f"\nDone! Found {len(all_models)} unique models total.")
````

````python
import os

# Create the folder if it doesn't exist
os.makedirs(, exist_ok=True)

# Build results dataframe and save
results = pd.DataFrame([
    {"Model": model, "# of Papers": count}
    for model, count in sorted(all_models.items(), key=lambda x: -x[1])
])

print(results.head(20))

results.to_excel(), index=False)
print("\nSaved!")
````
